```python
# Passo 1 -> Pegar o arquivo com as agencias do BB
import pandas as pd

data = pd.read_excel("202107-AGENCIAS BB.xlsx", sheet_name="Agencias BB")

#Converte a coluna "COD Agência" para string
data["COD AGÊNCIA"] = data["COD AGÊNCIA"].apply(str)

# Preenche a coluna com zeros a esquerda 
data['COD AGÊNCIA'] = data['COD AGÊNCIA'].apply(lambda x: x.zfill(4))
display(data)
```

```python
# Passo 2 -> Criar a função para calcular o DV

def calculadv(agencia):    
    # Calcula cada posição de cada caractere da agência
    
    p1 = int(agencia[3]) * 2
    p2 = int(agencia[2:3]) * 3
    p3 = int(agencia[1:2]) * 4
    p4 = int(agencia[0:1]) * 5
    
    # Soma os valores das posições
    Totalp = p1 + p2 + p3 + p4
      
    # Pegar o resto da divisão
    restDivisao = Totalp % 11
  
    # Subtrai de 11 o valor do resto da divisão
    dv = str(11 - restDivisao)
    
    # Se o DV for 10, trocar para "X"    
    if dv == "10": dv="X"
    
    return dv
```

```python
# Passo 3 -> atualizar a coluna DV 

#Percorre o indice com base no tamanho do dataframe
for i in range(len(data)):
    data.loc[i, "DV AGÊNCIA"] = calculadv(data.loc[i, "COD AGÊNCIA"])

display(data)
```

```python
#Passo4 -> Salvar o dataframe em novo arquivo
data.to_excel ("202107-AGENCIAS BB com DV.xlsx", index=False)
```
